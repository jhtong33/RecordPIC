```python
REQUIRED_EACH_CATEGORY = 4

# collect = dict()

with open('collect.json', 'r') as f:
    collect = json.loads(f.read())
```

```python
for i, (category, dfc) in enumerate(df.groupby('category')):
    item_type = dfc.item_type.iloc[0]
    
    print(i+1, '/', len(set(df.category)), end='\r')
    
    category_done = False
    candidates = list()
    
    if '其他' in category:
        continue
    
    if category in collect:
        continue
    
    item_names = list(set(
        
        dfc.item_name))
    random.shuffle(item_names)
    c_len = len(item_names)
    for j, item_name in enumerate(item_names):
        
        # if not '冷藏' in item_name:
        #     continue
        
        
        if '外送費' in item_name:
            continue
        if '外送員小費' in item_name:
            continue
        if '平台費' in item_name:
            continue

        # print()
        #y for match any for else: 
        varify = input(f'{j}/{c_len}, {item_type}-{category}, {item_name} ')
        
        # print(varify)
        
        insert_kw = 'insert'
        if varify.lower().replace(' ', '') == 'y' or varify.isdigit() or varify.startswith(insert_kw):
            
            if varify.isdigit():
                try:
                    
                    candidates.append(item_names[int(varify)])
                    print(f'get: {item_names[int(varify)]}, len(candidates): {len(candidates)}')
                except:
                    pass
                
            if varify.lower().replace(' ', '') == 'y':
                
                candidates.append(item_name)
                print(f'get: {item_name}, len(candidates): {len(candidates)}')
                
                
            if varify.startswith(insert_kw):
                
                _in = varify.replace(insert_kw,'')
                
                
                candidates.append(_in)
                print(f'get: {_in}, len(candidates): {len(candidates)}')
                
            collect[category] = candidates
            if len(candidates)< min(c_len, REQUIRED_EACH_CATEGORY):
                continue
                
            print(f'done: {category}, len(candidates): {len(candidates)}')

            category_done = True
            collect[category] = candidates

            
            
            
            with open('collect.json', 'w') as f:
                f.write(json.dumps(collect))

            break

            
            
            
```
