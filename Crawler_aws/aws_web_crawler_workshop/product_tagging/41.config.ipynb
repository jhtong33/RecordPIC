```python
import pandas as pd
df = pd.read_csv('data/samples.csv', index_col=0)
dfc = pd.read_csv('data/category.csv')
```

```python
config = dict()
config_catonly = dict()
for i, d in dfc.iterrows():
    
    if '其他' in d['category']:
        continue
    
    config[d['category']] = list(map(lambda x: x.strip(), filter(pd.notna, [d['item1'], d['item2']])))
    config_catonly[d['category']] = list()
    
config
```

```python
with open('workflow_asset/catogory.json', 'w') as f:
    f.write((json.dumps(config_catonly)))
```
