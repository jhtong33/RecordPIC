```python
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif'] = ['Taipei Sans TC Beta']
```

```python
fig, ax = plt.subplots(figsize=(10,8))
s=21
d=7
for i, (t, _df) in enumerate(df.groupby(by='item_type')):

    if not(s<i<=s+d):
        continue
        
    _df['item_name'].apply(len).hist(bins=10, log=False,label=t, alpha=0.3, density=True)
    # plt.show()
    # df['item_tokens'].apply(len).hist()

plt.legend(loc='best')
plt.show()
```
